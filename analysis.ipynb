```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
column_names = ["t","PGL3", "PGL3-mRNA", "PGL3p", "MEG3", "MEG3-mRNA", "MEG3p", "mRNA","MEX5","MEX5-mRNA"]
anterior = pd.read_csv('WithComp/anterior.csv',header=None, names=column_names,delimiter=' ')
posterior = pd.read_csv('WithComp/posterior.csv',header=None, names=column_names, delimiter=' ')
```

```python
anterior['scaled_PGL3-mRNA'] = anterior['PGL3-mRNA']/0.625
posterior['scaled_PGL3-mRNA'] = posterior['PGL3-mRNA']/0.375
anterior['scaled_MEG3-mRNA'] = anterior['MEG3-mRNA']/0.625
posterior['scaled_MEG3-mRNA'] = posterior['MEG3-mRNA']/0.375
```

```python
fig, ax = plt.subplots()
sns.set_palette('bright')
sns.lineplot(data=posterior, ax=ax,x='t',y='scaled_PGL3-mRNA',alpha=0.7,label='Posterior')
sns.lineplot(data=anterior, ax=ax, x='t',y='scaled_PGL3-mRNA',alpha=0.7,label='Anterior')
ax.set_xlabel('Time (sec)')
ax.set_ylabel('Relative PGL3-mRNA Concentration')
ax.set_title('Relative Posterior and Anterior Concentrations of PGL3-mRNA')
plt.legend(loc='best')
plt.show()
```

```python
fig, ax = plt.subplots()
sns.lineplot(data=posterior, ax=ax,x='t',y='scaled_MEG3-mRNA',alpha=0.7,label='Posterior')
sns.lineplot(data=anterior, ax=ax, x='t',y='scaled_MEG3-mRNA',alpha=0.7,label='Anterior')
ax.set_xlabel('Time (sec)')
ax.set_ylabel('Relative MEG3-mRNA Concentration')
ax.set_title('Relative Posterior and Anterior Concentrations of MEG3-mRNA')
plt.legend(loc='best')
plt.show()
```

```python
display(posterior['scaled_PGL3-mRNA'].mean())
display(anterior['scaled_PGL3-mRNA'].mean())
display(posterior['scaled_MEG3-mRNA'].mean())
display(anterior['scaled_MEG3-mRNA'].mean())
```

```python
fig, ax = plt.subplots()
sns.lineplot(data=posterior, ax=ax,x='t',y='MEX5-mRNA',alpha=0.7,label='Posterior')
sns.lineplot(data=anterior, ax=ax, x='t',y='MEX5-mRNA',alpha=0.7,label='Anterior')
ax.set_xlabel('Time (sec)')
ax.set_ylabel('Relative MEX5-mRNA Concentration')
ax.set_title('Relative Posterior and Anterior Concentrations of MEX5-mRNA')
plt.legend(loc='best')
plt.show()
```
