```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
column_names = ["t","PGL3", "PGL3-mRNA", "PGL3p", "MEG3", "MEG3-mRNA", "MEG3p", "mRNA","MEX5","MEX5-mRNA"]
anterior = pd.read_csv('WithComp/plkko_anterior.csv',header=None, names=column_names,delimiter=' ')
posterior = pd.read_csv('WithComp/plkko_posterior.csv',header=None, names=column_names, delimiter=' ')
```

```python
anterior['scaled_PGL3-mRNA'] = anterior['PGL3-mRNA']/0.625
posterior['scaled_PGL3-mRNA'] = posterior['PGL3-mRNA']/0.375
anterior['scaled_MEG3-mRNA'] = anterior['MEG3-mRNA']/0.625
posterior['scaled_MEG3-mRNA'] = posterior['MEG3-mRNA']/0.375
```

```python
fig, ax = plt.subplots()
sns.lineplot(data=posterior, ax=ax,x='t',y='scaled_PGL3-mRNA',color='red',alpha=0.7)
sns.lineplot(data=anterior, ax=ax, x='t',y='scaled_PGL3-mRNA',color='blue',alpha=0.7)
ax.set_xlabel('Time')
ax.set_ylabel('Relative PGL3-mRNA Concentration')
plt.savefig('WithComp/PGL3_PLK_KO.png',bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots()
sns.lineplot(data=posterior, ax=ax,x='t',y='scaled_MEG3-mRNA',color='red',alpha=0.7)
sns.lineplot(data=anterior, ax=ax, x='t',y='scaled_MEG3-mRNA',color='blue',alpha=0.7)
fig.savefig('WithComp/MEG3_PLK_KO.png')
```

```python
display(posterior['scaled_MEG3-mRNA'].mean())
display(anterior['scaled_MEG3-mRNA'].mean())
display(posterior['scaled_PGL3-mRNA'].mean())
display(anterior['scaled_PGL3-mRNA'].mean())
```

```python
fig, ax = plt.subplots()
sns.lineplot(data=posterior,ax=ax,x='t',y='MEX5-mRNA',color='red',alpha=0.7)
sns.lineplot(data=anterior, ax=ax, x='t',y='MEX5-mRNA',color='blue',alpha=0.7)
plt.savefig('WithComp/MEX5_PLK_KO.png')
```
