```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
column_names = ["t","PGL3", "PGL3-mRNA", "PGL3p", "MEG3", "MEG3-mRNA", "MEG3p", "mRNA","MEX5","MEX5-mRNA"]
anterior = pd.read_csv('anterior.csv',header=None, names=column_names,delimiter=' ')
posterior = pd.read_csv('posterior.csv',header=None, names=column_names, delimiter=' ')
```

```python
anterior['scaled_PGL3-mRNA'] = anterior['PGL3-mRNA']/0.625
posterior['scaled_PGL3-mRNA'] = posterior['PGL3-mRNA']/0.375
anterior['scaled_MEG3-mRNA'] = anterior['MEG3-mRNA']/0.625
posterior['scaled_MEG3-mRNA'] = posterior['MEG3-mRNA']/0.375
```

```python
fig, ax = plt.subplots()
sns.lineplot(data=posterior, ax=ax,x='t',y='scaled_PGL3-mRNA',color='red',alpha=0.7)
sns.lineplot(data=anterior, ax=ax, x='t',y='scaled_PGL3-mRNA',color='blue',alpha=0.7)
```
