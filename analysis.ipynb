```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
anterior = pd.read_csv('WithComp/anterior.csv')
posterior = pd.read_csv('WithComp/posterior.csv')

anterior['MEG3_fast'] = anterior['MEG3']+anterior['MEG3p']
posterior['MEG3_fast'] = posterior['MEG3']+posterior['MEG3p']
```

```python
anterior['scaled_PGL3-mRNA'] = anterior['PGL3-mRNA']/0.625
posterior['scaled_PGL3-mRNA'] = posterior['PGL3-mRNA']/0.375
anterior['scaled_PGL3'] = anterior['PGL3']/0.625
posterior['scaled_PGL3'] = posterior['PGL3']/0.375
anterior['scaled_MEG3p'] = anterior['MEG3p']/0.625
posterior['scaled_MEG3p'] = posterior['MEG3p']/0.375
anterior['scaled_MEG3_fast'] = anterior['MEG3_fast']/0.625
posterior['scaled_MEG3_fast'] = posterior['MEG3_fast']/0.375
anterior['scaled_MEG3-mRNA'] = anterior['MEG3-mRNA']/0.625
posterior['scaled_MEG3-mRNA'] = posterior['MEG3-mRNA']/0.375
#anterior['scaled_Pgran'] = anterior['Pgran']/0.625
#posterior['scaled_Pgran'] = posterior['Pgran']/0.375
```

```python
fig, ax = plt.subplots()
sns.lineplot(data=posterior, ax=ax,x='time',y='Pgran',alpha=0.7,label='Posterior')
sns.lineplot(data=anterior, ax=ax, x='time',y='Pgran',alpha=0.7,label='Anterior')
ax.set_xlabel('Time (sec)')
ax.set_ylabel('Relative P granule Concentration')
ax.set_title('Relative Posterior and Anterior Concentrations of P granules')
plt.legend(loc='best')
plt.show()
```
