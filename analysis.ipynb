```python
def normalise_cell(df):
    df[1] = df[1]/(3500/27000)
    df[2] = df[2]/(2208.3333/9000)
    df[3] = df[3]/(2250/9000)
    df[4] = df[4]/(2208.3333/9000)
    df[5] = df[5]/(3500/27000)
    return df
```

```python
meg3_loc = pd.DataFrame()
for i in range(10):
    meg3_loc_i = pd.read_csv(f'Pgran/intermed_runs/{i}_mex5_ko_megrna_spacecount.csv', header=None)
    meg3_loc = pd.concat([meg3_loc, meg3_loc_i])
meg3_loc = normalise_cell(meg3_loc)
```

```python
import numpy as np

# Constants
cylinder_radius = 15  # radius of the cylinder
cylinder_length = 20  # length of the cylinder
hemisphere_radius = 15  # radius of the hemisphere

# Number of boxes along each axis
long_axis_boxes = 10
short_axis_boxes = 3

# Calculate box dimensions
box_length = cylinder_length / long_axis_boxes
box_width = 2 * hemisphere_radius / short_axis_boxes

# Populate the counts in each box (replace this with your actual molecule counts)
# For demonstration purposes, using random counts
boxes = np.genfromtxt('Pgran/Pgran_spacecount.csv',delimiter=',')
boxes1 = np.genfromtxt('Pgran/megrna_spacecount.csv', delimiter=',')

# Scale counts based on box area
def scale_counts(boxes):
    for i in range(long_axis_boxes):
        for j in range(short_axis_boxes):
            # Calculate box area
            box_area = box_length * box_width
            
            # Calculate box volume using the height dimension
            box_volume = box_area * box_width
            
            # Scale counts based on volume
            boxes[j, i] = boxes[j, i] / box_volume

P_gran = scale_counts(boxes)
megrna = scale_counts(boxes1)
fig, ax = plt.subplots(2,1,figsize=(8,6))
im1 = ax[0].imshow(megrna, cmap='viridis',interpolation='sinc')
ax[0].set_title('MEG3-mRNA Localisation')
ax[0].set_xticks([0,2,4,6,8,9],['Posterior','','','','','Anterior'])
ax[0].set_xlabel('Position')
ax[0].set_yticks([0,1,2],['','',''])
im2 = ax[1].imshow(P_gran, cmap='viridis', interpolation='sinc')
ax[1].set_title('P Granule Localisation')
cbar = fig.colorbar(im2, ax=ax, orientation='vertical', fraction=0.05, pad=0.1)
cbar.set_label('Relative Concentration',size=12)
plt.xticks([0,2,4,6,8,9],['Posterior','','','','','Anterior'])
plt.xlabel('Position')
ax[1].set_yticks([0,1,2],['','',''])
ax[1].set_title('P Granule Localisation')
```

```python
def process_dfs(condition, folder=None):
    if folder:
        grad_list = []
        for i in range(10):
            posterior_i = pd.read_csv(f'{folder}/{i}_{condition}_posterior.csv')
            posterior_mean_i = posterior_i.mean()
            anterior_i = pd.read_csv(f'{folder}/{i}_{condition}_anterior.csv')
            anterior_mean_i = anterior_i.mean()
            grad = 1-(anterior_mean_i['Pgran']/posterior_mean_i['Pgran'])
            grad_list.append(grad)
        grad_df = pd.DataFrame({'Absolute Gradient':grad_list,'Condition':[condition for i in range(10)]})
    return grad_df
```

```python
def process_df_plk(condition, folder=None):
    if folder:
        grad_list = []
        for i in range(10):
            posterior_i = pd.read_csv(f'{folder}/{i}_{condition}posterior.csv')
            posterior_mean_i = posterior_i.mean()
            anterior_i = pd.read_csv(f'{folder}/{i}_{condition}anterior.csv')
            anterior_mean_i = anterior_i.mean()
            grad = 1-(anterior_mean_i['Pgran']/posterior_mean_i['Pgran'])
            grad_list.append(grad)
        grad_df = pd.DataFrame({'Absolute Gradient':grad_list,'Condition':[condition for i in range(10)]})
    return grad_df
```

```python
wt = process_df_plk('','Pgran')
wt['Condition'] = 'WT'
meg3 = process_dfs('meg_ko','Pgran/intermed_runs')
meg3['Condition'] = 'meg-3 KO'
plk1 = process_df_plk('','Pgran/smol_data')
plk1['Condition'] = 'plk-1 KO'
pgl = process_dfs('pgl3_ko','Pgran/intermed_runs')
pgl['Condition'] = 'pgl-3 KO'
mex5 = process_dfs('mex5_ko','Pgran/intermed_runs')
mex5['Condition'] = 'mex-5 KO'
df_to_plot = pd.concat([wt, plk1, pgl, mex5, meg3], ignore_index=True)
```

```python
fig = sns.catplot(data=df_to_plot, x='Condition',y='Absolute Gradient', legend='full', kind='box')
fig.ax.set_ylabel('P Granule Absolute Gradient')
fig.ax.set_xlabel('Perturbation')
fig.figure.suptitle('P Granule Gradients with Different Knockouts')
plt.tight_layout()
plt.savefig('plots/P_granule_gradients.pdf')
```
