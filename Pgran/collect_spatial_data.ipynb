```python
import smoldyn
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def get_grads(anterior, posterior):
    grads_to_calc = ['scaled_MEG3', 'scaled_MEG3p', 'scaled_MEG3-mRNA', 'scaled_MEG3_fast', 'scaled_MEG3_total']
    grad_list=[]
    for species in grads_to_calc:
        grad = 1-(anterior[species].mean()/posterior[species].mean())
        grad_list.append(grad)
    return grad_list
```

```python
s = smoldyn.Simulation.fromFile('elegans_with_competition.txt')
```

```python
def runSims(num_sims):
    data = pd.DataFrame()
    for i in range(num_sims):
        s.run(5000,5,overwrite=True)
        megrna = pd.read_csv('megrna_spacecount.csv')
        data = pd.concat([data, megrna], ignore_index=True)
        data.to_csv('wild_type_10_sims.csv')
    return data


```

```python
data = runSims(1200)
data.to_csv('ml_train_data.csv')
```
